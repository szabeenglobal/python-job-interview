Using Historical Volatility To Gauge Future Risk

        <li> ARCH (Autoregressive Conditional Heteroskedasticity) </li>

```python
import datetime
import time

# Closing Price: The closing price is the raw price, which is just the cash value of the last transacted price before the market closes. 
# Adjusted Closing Price: The adjusted closing price factors in anything that might affect the stock price after the market closes. 
# A stock's price is typically affected by supply and demand of market participants.
import matplotlib.pyplot as plt
stocks = 'AAPL'
start = datetime.datetime(2019, 3, 15)
end = datetime.datetime(2022, 5, 3)

apple1 = yf.download(stocks, start=start, end = end, interval='1d') # daily interval stock price

# print(apple1.boxplot)
# calculate return
# "Pandas dataframe. pct_change() function calculates the percentage change between the current and a prior element. 
# This function by default calculates the percentage change from the immediately previous row. Note : 
# This function is mostly useful in the time-series data"
ret = 100 * (apple1.pct_change()[1:]['Adj Close'])

# calculate the 5-da rolling window realized volatility
realized_vol = ret.rolling(5).std() #rolling window 5 days

plt.figure(figsize=(10, 6))
plt.plot(realized_vol.index,realized_vol)
plt.title('Realized Volatility- APPL')
plt.ylabel('Volatility')
plt.xlabel('Date')
plt.show()
```

## Support Vector Regression: GARCH

Support vector machine (SVM) is a supervised learning algorithm  
the variance is the average of the squared returns

```python
# import the dependecies

from sklearn.svm import SVR
from scipy.stats import uniform as sp_rand
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error as mse




# convert it in a pandas dataframe
realized_vol = pd.DataFrame(realized_vol)

realized_vol.reset_index(drop=True, inplace=True)
# realized_vol.info()

# calculate square of historical return
returns_svm = ret ** 2  #variance is the average of the squared returns
returns_svm = returns_svm.reset_index()
del returns_svm['Date']

X = pd.concat([realized_vol, returns_svm], axis=1, ignore_index=True)
X.head() # check for null values, only first four rows are null


X = X[4:].copy()  #get rid of first 4 rows
X = X.reset_index()
X.drop('index', axis=1, inplace=True)

realized_vol = realized_vol.dropna().reset_index()
realized_vol.drop('index', axis=1, inplace=True)


svr_lin = SVR(kernel='linear') 
svr_poly = SVR(kernel='poly', degree=2) 
svr_rbf = SVR(kernel='rbf')  #Radial Basis Function (RBF) kernel SVM.
```

```python
#clf.fit(X = realized volatility, Y = returns sum])
# for hyperparameter tuning use RandomizedSearchCV
# C is cost
# low gamma loosly fit - high gamma perfect fit
# for continuous data 
para_grid = {   'gamma': ['auto', 1e-3, 1e-4],
                'C': [0.01, .1, 1, 10, 100, 1000],
                'epsilon': sp_rand()}  

# find the best parameters
clf = RandomizedSearchCV(svr_lin, para_grid) 
#fit the svr model
clf.fit(X, realized_vol) 

#make predictions
predict_svr_lin = clf.predict() 

predict_svr_lin = pd.DataFrame(predict_svr_lin)

#Calculate error

rmse_svr = np.sqrt(mse(realized_vol.iloc[] / 100,
                    predict_svr_lin / 100))
print('The RMSE value of SVR with Linear Kernel is {:.6f}'
    .format(rmse_svr))




```
